```python
import matplotlib.pyplot as plt
from DispersionRelation import EquatorialWaveDispersion
ew = EquatorialWaveDispersion(
    nPlanetaryWave=200,
    rlat=0.0,
    Ahe=[50.0, 25.0, 12.0],
)

Afreq, Apzwn, wave_names = ew.compute(waves="all")

# choose one equivalent depth to plot
ie = 0  # index of he in ew.Ahe
s0 = Apzwn[0, 0, :]
s1 = Apzwn[0, 1, :]
s2 = Apzwn[0, 2, :]

plt.figure(figsize=(9, 6))

for iw, wname in enumerate(wave_names):
    plt.plot(s1, Afreq[iw, 0, :], label=wname)
    plt.plot(s2, Afreq[iw, 1, :], label=wname)
    plt.plot(s0, Afreq[iw, 2, :], label=wname)

plt.axhline(0, color="k", linewidth=0.6)
plt.axvline(0, color="k", linewidth=0.6)
plt.xlabel("Zonal wavenumber index  s")
plt.ylabel("Frequency  (cycles/day)")
plt.title(f"Equatorial dispersion (he = {ew.Ahe[ie]:.1f} m)")
plt.grid(True, linestyle="--", alpha=0.4)
plt.legend()
plt.show()
```
